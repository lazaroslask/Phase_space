# 3D Plot

---



```python
import matplotlib.pyplot as plt
import numpy as np

# -------- First plot --------
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # correct way to add 3D axes

z1 = np.linspace(0, 1, 100)
x1 = z1 * np.sin(25 * z1)
y1 = z1 * np.cos(25 * z1)

ax.plot3D(x1, y1, z1, 'green')
ax.set_title("3D Line Plot")
plt.show()

# -------- Second plot --------
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')  # create new 3D axes in fig2

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.cos(x*y)

ax2.plot_surface(x, y, z, cmap='viridis')
ax2.set_title("3D Surface Plot")
plt.show()
```

# Symbolic

```python
from sympy import *

x = Symbol('x')
expr = sin(x) + x**2 + 4*x

print("expr = ", expr)
print("diff = ", diff(expr))
print("integrate = ", integrate(expr))
```

# Neural Networks with Python

```python
inputs = [1,2,3]
weights = [ 0.2 , 0.8 , - 0.5 ]
bias = 2
output = (inputs[ 0 ] * weights[ 0 ] + inputs[ 1 ] * weights[ 1 ] + inputs[ 2 ] * weights[ 2 ] + bias)

print(output)
```
